```python
data.drop(['instance_id', 'artist_name', 'track_name', 'duration_ms', 'key', 'mode', 'obtained_date', 'valence', 'liveness' ], axis=1, inplace=True)
data.dtypes
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split

# Assuming 'data' is your original dataframe

# Separate features and target
y = data[['music_genre']]
X = data.drop(['music_genre'], axis=1)

# Convert 'tempo' to numeric, handling errors
X['tempo'] = pd.to_numeric(X['tempo'], errors='coerce')

# Drop rows with NaN values in 'tempo' to ensure X and y have the same number of samples
X.dropna(subset=['tempo'], inplace=True)  # Drop rows with NaN in 'tempo'
y = y.iloc[X.index]  # Align y with the updated X index

# Proceed with scaling and encoding
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

le = LabelEncoder()
y = le.fit_transform(y)
```

```python
X.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=123)
```

```python
from xgboost import XGBClassifier
xgb= XGBClassifier(n_estimators=101,random_state=123,learning_rate=0.1)
xgb.fit(X_train,y_train)
y_pred = xgb.predict(X_test)
xgb.score(X_test,y_test)
```
